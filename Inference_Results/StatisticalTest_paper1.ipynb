```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from scipy.stats import mannwhitneyu,kruskal,ranksums,wilcoxon
import scipy
from scikit_posthocs import posthoc_conover
from cliffs_delta import cliffs_delta
def kw_test():
  data=pd.read_csv('Inference_Results_Paper1.csv')
  df = pd.DataFrame()
  bonferroni_results={'large':'L','medium':'M','small':'S','negligible':'N'}
  models=["ResNet","ResNext","FCN","DUC"]
  for model in models:
      pvalues=[]
      groups_names=["Mobile","Edge","Cloud"]
      data1=data[(data['Device']=="Mobile") & (data['Model']==model)]['Inference']
      data2=data[(data['Device']=="Edge") & (data['Model']==model)]['Inference']
      data3=data[(data['Device']=="Cloud") & (data['Model']==model)]['Inference']
      k_result = kruskal(data1,data2,data3)
      print(model,' Identity (Mobile, Edge, Cloud) ',k_result, "p < 0.05:", k_result.pvalue < 0.05)
      k_pvalue=str(k_result.pvalue)
      adjp_delta_values=[]
      if 'e' in k_pvalue:
          k_pvalue=k_pvalue[:3]+'e'+k_pvalue.split('e')[1]
      else:
          k_pvalue=get_number_until_nonzero(k_pvalue)
      if float(k_pvalue) < 0.05:
          adjp_delta_values.append(k_pvalue)
      if float(k_pvalue) >= 0.05:
          continue
      groups=[data1,data2,data3]
      pairwise_results = posthoc_conover(groups).values
      for i in range(len(groups)):
          for j in range(i+1, len(groups)):
              d=cliffs_delta(groups[i], groups[j])
              p=pairwise_results[i][j]
              print(groups_names[i],' - ',groups_names[j],":", ' adjusted p-value = ',str(p), ', rejected = ', str(p < 0.05), ', Cliffs Delta = ',str(d))
              p=str(p)
              if 'e' in p:
                   p=p[:3]+'e'+p.split('e')[1]
              else:
                   p=get_number_until_nonzero(p)
              if float(p) < 0.05:
                  adjp_delta_values.append(p+'('+ str("{:.2f}".format(d[0]))+':'+bonferroni_results[d[1]]+')')
              if float(p) >= 0.05:
                  adjp_delta_values.append('')
      column=pd.Series(adjp_delta_values, index=['Mobile(I),Edge(I),Cloud(I)','Mobile(I),Edge(I)','Mobile(I),Cloud(I)',
                                                 'Edge(I),Cloud(I)'])
      df[model] = column
  return df
kw_test()
```
