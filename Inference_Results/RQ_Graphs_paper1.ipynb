```python
import numpy as np
import pandas as pd
import scipy
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
mean_values=[]
median_values=[]
subjects=["ResNet","ResNext","FCN","DUC"]
devices=["Mobile","Edge","Cloud","Mobile-Edge","Mobile-Cloud","Edge-Cloud"]
data=pd.read_csv('Inference_Results_Paper1 - Copy.csv')
data['Inference']=np.log(data['Inference'])
accuracy=pd.read_csv('accuracy.csv')
df = pd.DataFrame(columns=['Inference Latency(s)','Top5(%)', 'Top1(%)','mIOU(%)','Device', 'Model','Subject'])
for subject in subjects:
    models=[subject,subject+" EarlyExit", subject+ " Quantize", subject+ " Partition", subject+" Quantize EarlyExit", subject+" Quantize EarlyExit Partition"]
    for i in range(len(devices)):
        for j in range(len(models)):
            median=np.median(data[(data['Device']==devices[i]) & (data['Model']==models[j])]['Inference'])
            if subject in ["FCN","DUC"]:
                miou=np.max(accuracy[accuracy['Model']==models[j]]['miou'])
                df = df._append({'Inference Latency(s)': median, 'Top5(%)':np.nan,'Top1(%)':np.nan,'mIOU(%)':miou,'Device': devices[i],'Model': models[j],'Subject':subject}, ignore_index=True)
            if subject in ["ResNet","ResNext"]:
                top5=np.max(accuracy[accuracy['Model']==models[j]]['top5'])
                top1=np.max(accuracy[accuracy['Model']==models[j]]['top1'])
                df = df._append({'Inference Latency(s)': median, 'Top5(%)':top5,'Top1(%)':top1,'mIOU(%)':np.nan,'Device': devices[i],'Model': models[j],'Subject':subject}, ignore_index=True)
df['Model_Device'] = df['Device'] + " - " + df['Model']
```
