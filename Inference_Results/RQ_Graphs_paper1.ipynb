```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
dataframe = pd.read_csv('Inference_Results_Paper1.csv')
```

```python
dataframe['Inference']=np.log(dataframe['Inference'])
```

```python
fig, axs = plt.subplots(2, 2, tight_layout=True)
ymax=math.ceil(dataframe[((dataframe['Model']=="ResNet")|(dataframe['Model']=="ResNext")|(dataframe['Model']=="FCN")|(dataframe['Model']=="DUC")) & ((dataframe['Device']=="Mobile") | (dataframe['Device']=="Edge") | (dataframe['Device']=="Cloud"))]['Inference'].max())
sns.boxplot(data=dataframe[(dataframe['Model']=="ResNet") & ((dataframe['Device']=="Mobile") | (dataframe['Device']=="Edge") | (dataframe['Device']=="Cloud"))],x="Device",y="Inference", ax=axs[0][0], linewidth=0.8)
axs[0][0].set_title('ResNet')
axs[0][0].set_ylabel('log(s)')
axs[0][0].set_ylim(([0, ymax]))
sns.boxplot(data=dataframe[(dataframe['Model']=="ResNext") & ((dataframe['Device']=="Mobile") | (dataframe['Device']=="Edge") | (dataframe['Device']=="Cloud"))],x="Device",y="Inference", ax=axs[0][1], linewidth=0.8)
axs[0][1].set_title('ResNext')
axs[0][1].set_ylabel('log(s)')
axs[0][1].set_ylim(([0, ymax]))
sns.boxplot(data=dataframe[(dataframe['Model']=="FCN") & ((dataframe['Device']=="Mobile") | (dataframe['Device']=="Edge") | (dataframe['Device']=="Cloud"))],x="Device",y="Inference", ax=axs[1][0], linewidth=0.8)
axs[1][0].set_title('FCN')
axs[1][0].set_ylabel('log(s)')
axs[1][0].set_ylim(([0, ymax]))
sns.boxplot(data=dataframe[(dataframe['Model']=="DUC") & ((dataframe['Device']=="Mobile") | (dataframe['Device']=="Edge") | (dataframe['Device']=="Cloud"))],x="Device",y="Inference", ax=axs[1][1], linewidth=0.8)
axs[1][1].set_title('DUC')
axs[1][1].set_ylabel('log(s)')
axs[1][1].set_ylim(([0, ymax]))
plt.show()
fig.savefig('img_rq1_results.pdf', bbox_inches='tight',format='pdf')
```
